```python
import pandas as pd
```

```python
file = "Resources/purchase_data.csv"
```

```python
df = pd.read_csv(file)
df.head()
```

```python
# removing duplicate players
noDuplicate_df = df.drop_duplicates(subset="SN")
noDuplicate_df.head()
```

```python
# count total number of players
player_df = noDuplicate_df["SN"].count()

# display total number of players
total_player = pd.DataFrame({"Total Players": [player_df]})
total_player
```

```python
# count unique item
unique_item_df = df['Item Name'].nunique()

# count number of purchases
purchase_df = df['Purchase ID'].nunique()

# count average price
avg_price_df = round(df['Price'].mean(), 2)

# count total revenue
revenue_df = df["Price"].sum()

# summary dataframe
summary_df = pd.DataFrame({
    "Number of Unique Items": [unique_item_df],
    "Average Price": "$" + str(avg_price_df),
    "Number of Purchases": [purchase_df],
    "Total Revenue": "$"+ str(revenue_df)
})
summary_df
```

```python
# grouping by gender
grouped_gender = noDuplicate_df.groupby(["Gender"])

# count male, female, and other
gender_df = noDuplicate_df["Gender"].value_counts()

# calculate gender percentage
gender_percent = gender_df/player_df*100

# demography dataframe
demographic_df = pd.DataFrame({"Total Count": gender_df,
                                "Percentage of Players": gender_percent.map("{:.2f}%".format)})
demographic_df
```

```python
# setting groupby with duplicates
grouped_df = df.groupby(["Gender"])

# purchase count
purchase_count = grouped_df["Price"].count()

# average purchase price
avg_purchase_price = grouped_df["Price"].mean()

# total purchase value
total_purchase_value = grouped_df["Price"].sum()

# average total purchase per person
avg_pperson = total_purchase_value/gender_df


# purchase analysis dataframe
purchase_analysis_df = pd.DataFrame({"Purchase Count": purchase_count,
                                    "Avg Purchase Price": avg_purchase_price.map("${:.2f}".format),
                                    "Total Purchase Value": total_purchase_value.map("${:.2f}".format),
                                    "Avg Total Purchase per Person": avg_pperson.map("${:.2f}".format)})
purchase_analysis_df
```

```python
# creating bins
bins = [0, 9, 14, 19, 24, 29, 34, 39, 45]

# create labels for these bins
group_labels = "<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "45+"
```

```python
# slice the data and place it into bins
age_range1 = pd.cut(noDuplicate_df["Age"], bins, labels=group_labels)
age_range2 = pd.cut(df["Age"], bins, labels=group_labels)

# add age range to noduplicate dataframe
noDuplicate_df["Age Range"] = age_range1

# add age range to original dataframe
df["Age Range"] = age_range2
```

```python
# grouping by age range
grouped_age = noDuplicate_df.groupby("Age Range")

# total count of age range
total_count_age = grouped_age["Age"].count()

# percent of players
player_percent = total_count_age/player_df*100

# age demographic dataframe
age_dem_df = pd.DataFrame({"Total Count": total_count_age,
                            "Percentage of Players": player_percent.map("{:.2f}%".format)})
age_dem_df
```

```python
# setting groupby with duplicates
grouped_age_df = df.groupby(["Age Range"])

# purchase count
age_purchase_count = grouped_age_df["Price"].count()

# average purchase price
age_avg_purchase_price = grouped_age_df["Price"].mean()

# total purchase value
age_total_purchase_value = grouped_age_df["Price"].sum()

# average total purchase per person
age_avg_pperson = age_total_purchase_value/total_count_age


# age purchase analysis dataframe
age_purchase_analysis_df = pd.DataFrame({"Purchase Count": age_purchase_count,
                                    "Avg Purchase Price": age_avg_purchase_price.map("${:.2f}".format),
                                    "Total Purchase Value": age_total_purchase_value.map("${:.2f}".format),
                                    "Avg Total Purchase per Person": age_avg_pperson.map("${:.2f}".format)})
age_purchase_analysis_df
```

```python
# group by SN
grouped_sn = df.groupby(["SN"])

# purchase count
sn_purchase_count = grouped_sn["Price"].count()

# average purchase price
sn_avg_purchase_price = grouped_sn["Price"].mean()

# total purchase value
sn_total_purchase_value = grouped_sn["Price"].sum()

# sn purchase analysis dataframe
sn_purchase_analysis_df = pd.DataFrame({"Purchase Count": sn_purchase_count,
                                    "Avg Purchase Price": sn_avg_purchase_price.map("${:.2f}".format),
                                    "Total Purchase Value": sn_total_purchase_value})
  
purchase_df = sn_purchase_analysis_df.sort_values("Total Purchase Value", ascending=False)

# formatting
purchase_df["Total Purchase Value"] = purchase_df["Total Purchase Value"].map("${:.2f}".format)

purchase_df.head()
```

```python
# group by item ID & item name
grouped_item = df.groupby(["Item ID", "Item Name"])

# purchase count
item_purchase_count = grouped_item["Price"].count()

# item price
item_price = grouped_item["Price"].unique()

# total purchase value
item_total_purchase_value = grouped_item["Price"].sum()

# sn purchase analysis dataframe
popular_analysis_df = pd.DataFrame({"Purchase Count": item_purchase_count,
                                    "Item Price": item_price.str[0].map("${:.2f}".format),
                                    "Total Purchase Value": item_total_purchase_value.map("${:.2f}".format)})
  
popular_analysis_df.sort_values("Purchase Count", ascending=False).head()
```

```python
# group by item ID & item name
grouped_item = df.groupby(["Item ID", "Item Name"])

# purchase count
item_purchase_count = grouped_item["Price"].count()

# item price
item_price = grouped_item["Price"].unique()

# total purchase value
item_total_purchase_value = grouped_item["Price"].sum()

# sn purchase analysis dataframe
profitable_analysis_df = pd.DataFrame({"Purchase Count": item_purchase_count,
                                    "Item Price": item_price.str[0].map("${:.2f}".format),
                                    "Total Purchase Value": item_total_purchase_value})

# formatting 
profitable_df = profitable_analysis_df.sort_values("Total Purchase Value", ascending=False)

profitable_df["Total Purchase Value"] = profitable_df["Total Purchase Value"].map("${:.2f}".format)

profitable_df.head()
```
