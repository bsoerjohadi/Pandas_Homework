```python
import pandas as pd
```

```python
school_data = "Resources/schools_complete.csv"
student_data = "Resources/students_complete.csv"
```

```python
school_df = pd.read_csv(school_data)
student_df = pd.read_csv(student_data)
```

```python
merged_df = pd.merge(student_df, school_df, how="left", on=["school_name", "school_name"])
merged_df.head()
```

```python
# total schools 
total_school = merged_df["school_name"].nunique()

# total students
total_students = merged_df["student_name"].count()
# total budget
total_budget = merged_df["budget"].unique().sum()

# avg math score
avg_math = merged_df["math_score"].mean()

# avg reading score
avg_reading = merged_df["reading_score"].mean()

# percentage of students with a passing math score (70 or greater)
pass_math = merged_df.loc[merged_df["math_score"] >= 70, :]
percent_math = pass_math["math_score"].count()/merged_df["math_score"].count()*100

# percentage of students with a passing reading score (70 or greater)
pass_reading = merged_df.loc[merged_df["reading_score"] >= 70, :]
percent_reading = pass_reading["reading_score"].count()/merged_df["reading_score"].count()*100


# percentage of students who passed math and reading (% Overall Passing)
pass_both = merged_df.loc[(merged_df["math_score"] >= 70) & (merged_df["reading_score"] >= 70)]
percent_both = pass_both["math_score"].count()/total_students*100

# create dataframe
summary_df = pd.DataFrame({"Total Schools": [total_school], "Total Students": [total_students], "Total Budgets": [total_budget],
                          "Avg Math Score": [avg_math], "Avg Reading Score": [avg_reading], "% Passing Math": [percent_math],
                          "% Passing Reading": [percent_reading], "% Overall Passing": [percent_both]})

summary_df["Total Students"] = summary_df["Total Students"].map("{:,}".format)
summary_df["Total Budgets"] = summary_df["Total Budgets"].map("${:,}".format)

summary_df
```

```python
# group by school name
grouped_school = merged_df.groupby(["school_name"])

# school type
school_type = grouped_school["type"].unique()

# total students per school
school_student = grouped_school["student_name"].count()

# total school budget
school_budget = grouped_school["budget"].unique()

# per student budget 
student_budget = school_budget/school_student

# avg math score per school 
school_avg_math = grouped_school["math_score"].mean()

# avg reading score per school
school_avg_reading = grouped_school["reading_score"].mean()

# percentage of students with a passing math score (70 or greater) per school
school_pass_math = merged_df.loc[merged_df["math_score"] >= 70, :].groupby(["school_name"])
school_percent_math = school_pass_math["math_score"].count()/grouped_school["math_score"].count()*100

# percentage of students with a passing reading score (70 or greater) per school
school_pass_reading = merged_df.loc[merged_df["reading_score"] >= 70, :].groupby(["school_name"])
school_percent_reading = school_pass_reading["reading_score"].count()/grouped_school["reading_score"].count()*100

# percentage of students who passed math and reading (% Overall Passing) per school
school_pass_both = merged_df.loc[(merged_df["math_score"] >= 70) & (merged_df["reading_score"] >= 70)].groupby(["school_name"])
school_percent_both = school_pass_both["math_score"].count()/school_student*100

# create dataframe
school_summary_df = pd.DataFrame({"School Type": school_type.str[0], "Total Students": school_student, "Total School Budget": school_budget.str[0], 
                                 "Per Student Budget": student_budget.str[0], "Avg Math Score": school_avg_math, "Avg Reading Score": school_avg_reading,
                                 "% Passing Math": school_percent_math, "% Passing Reading": school_percent_reading, "% Overall Passing": school_percent_both})

school_summary_df["Total Students"] = school_summary_df["Total Students"].map("{:,}".format)
school_summary_df["Total School Budget"] = school_summary_df["Total School Budget"].map("${:,}".format)

school_summary_df.style.format({"Per Student Budget": "${:,}"})
```

```python
# top five performing schools by % overall passing
ss1_df = school_summary_df.sort_values("% Overall Passing", ascending=False).head()
ss1_df.style.format({"Per Student Budget": "${:,}"})
```

```python
# five worst-performing schools by % overall passing
ss2_df = school_summary_df.sort_values("% Overall Passing", ascending=True).head()
ss2_df.style.format({"Per Student Budget": "${:,}"})
```

```python
# extracting data grades grouped by school name 
ninth = merged_df.loc[merged_df["grade"] == "9th", :].groupby(["school_name"])
tenth = merged_df.loc[merged_df["grade"] == "10th", :].groupby(["school_name"])
eleventh = merged_df.loc[merged_df["grade"] == "11th", :].groupby(["school_name"])
twelfth = merged_df.loc[merged_df["grade"] == "12th", :].groupby(["school_name"])

# average Math Score for students of each grade level (9th, 10th, 11th, 12th) at each school
ninth_math = ninth["math_score"].mean()
tenth_math = tenth["math_score"].mean()
eleventh_math = eleventh["math_score"].mean()
twelfth_math = twelfth["math_score"].mean()

# create dataframe
summary_school_math_df = pd.DataFrame({"9th": ninth_math, "10th": tenth_math, "11th": eleventh_math, "12th": twelfth_math})

summary_school_math_df
```

```python
# average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school
ninth_reading = ninth["reading_score"].mean()
tenth_reading = tenth["reading_score"].mean()
eleventh_reading = eleventh["reading_score"].mean()
twelfth_reading = twelfth["reading_score"].mean()

# create dataframe
summary_school_reading_df = pd.DataFrame({"9th": ninth_reading, "10th": tenth_reading, "11th": eleventh_reading, "12th": twelfth_reading})

summary_school_reading_df
```

```python
# break down school performances based on average Spending Ranges per student
# creating bins
bins = [0, 584, 629, 644, 675] 

# creating names for bins
group_names = ["<$584", "$585-629", "$630-644", "$645-675"]

x = school_summary_df.copy()
x["Spending Ranges (Per Student)"] = pd.cut(school_summary_df["Per Student Budget"] , bins, labels=group_names)
spending_range_df = x.groupby(x["Spending Ranges (Per Student)"]).mean()[["Avg Math Score", "Avg Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]]

spending_range_df["Avg Math Score"] = spending_range_df["Avg Math Score"].map("{:.2f}".format)
spending_range_df["Avg Reading Score"] = spending_range_df["Avg Reading Score"].map("{:.2f}".format)
spending_range_df["% Passing Math"] = spending_range_df["% Passing Math"].map("{:.2f}%".format)
spending_range_df["% Passing Reading"] = spending_range_df["% Passing Reading"].map("{:.2f}%".format)
spending_range_df["% Overall Passing"] = spending_range_df["% Overall Passing"].map("{:.2f}%".format)

spending_range_df
```

```python
# adding size to school summary dataframe 
size = grouped_school["size"].unique()
size_school_summary_df = pd.DataFrame({"School Type": school_type.str[0], "Size": size.str[0], "Total Students": school_student, "Total School Budget": school_budget.str[0], 
                                 "Per Student Budget": student_budget.str[0], "Avg Math Score": school_avg_math, "Avg Reading Score": school_avg_reading,
                                 "% Passing Math": school_percent_math, "% Passing Reading": school_percent_reading, "% Overall Passing": school_percent_both})
```

```python
# break down school performances based on school size
# creating bins
size_bins = [0, 1000, 2000, 5000]

# creating names for bins
size_group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

y = size_school_summary_df.copy()
y["School Size"] = pd.cut(size_school_summary_df["Size"], bins= size_bins, labels= size_group_names)
size_df = y.groupby(y["School Size"]).mean()[["Avg Math Score", "Avg Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]]

size_df
```

```python
# break down school performances based on school type
# group by school type
grouped_type = merged_df.groupby(["type"])

# total students per school
type_student = grouped_type["student_name"].count()

# avg math score per school type 
type_avg_math = grouped_type["math_score"].mean()

# avg reading score per school type
type_avg_reading = grouped_type["reading_score"].mean()

# percentage of students with a passing math score (70 or greater) per school type
type_pass_math = merged_df.loc[merged_df["math_score"] >= 70, :].groupby(["type"])
type_percent_math = type_pass_math["math_score"].count()/grouped_type["type"].count()*100

# percentage of students with a passing reading score (70 or greater) per school
type_pass_reading = merged_df.loc[merged_df["reading_score"] >= 70, :].groupby(["type"])
type_percent_reading = type_pass_reading["reading_score"].count()/grouped_type["type"].count()*100

# percentage of students who passed math and reading (% Overall Passing) per school
type_pass_both = merged_df.loc[(merged_df["math_score"] >= 70) & (merged_df["reading_score"] >= 70)].groupby(["type"])
type_percent_both = type_pass_both["math_score"].count()/type_student*100

# create dataframe
type_summary_df = pd.DataFrame({"Avg Math Score": type_avg_math, "Avg Reading Score": type_avg_reading,
                                 "% Passing Math": type_percent_math, "% Passing Reading": type_percent_reading, "% Overall Passing": type_percent_both})
type_summary_df
```
